````python
//DANGLING ELSE AND COMPARING FLOAT NUMBERS USING EPSILON AND FABS()
/*
üòµ Dangling else:
   - Happens when you write nested `if` without proper braces `{}`.
   - The compiler always associates the `else` with the nearest unmatched `if`.

üõë BAD (Ambiguous):
    if (x > 0)
        if (x < 10)
            printf("Between 1 and 9");
        else
            printf("Not greater than 0");

‚úÖ GOOD (Clear and intended):
    if (x > 0) {
        if (x < 10)
            printf("Between 1 and 9");
    } else {
        printf("Not greater than 0");
    }
*/
/*
üéØ Problem: Floating-point values can't be compared using `==` due to precision errors.
üìå Solution: Use an acceptable difference range (called EPSILON) and `fabs()` from <math.h>

    Example: To check if a and b are "almost equal":
        if (fabs(a - b) < EPSILON)

    - `fabs()` gives absolute value of float/double.
    - `EPSILON` is a small value like 0.000001
*/

#include <stdio.h>
#include <math.h>

#define EPSILON 0.000001

int main() {
    float a = 0.1f * 3;
    float b = 0.3f;

    printf("a = %.20f\n", a);  // just to show difference
    printf("b = %.20f\n", b);

    if (fabs(a - b) < EPSILON) {
        printf("‚úÖ a and b are approximately equal.\n");
    } else {
        printf("‚ùå a and b are NOT equal.\n");
    }

    return 0;
}

````

````python
/*------------------------------------------------------------
  üìò SWITCH-CASE STATEMENT IN C ‚Äî Full Explanation
-------------------------------------------------------------

üìå Purpose:
   ‚û§ Used for multi-way branching when comparing a variable to many constant values.
   ‚û§ Works **only with integral or enumerated types** (int, char, enum ‚Äî not float/double).

------------------------------------------------------------
üìö COMPARISON: switch-case vs if-else ladder

  if-else ladder:                            switch-case:
  ----------------                          ------------------
  ‚û§ Supports ranges, logic ops              ‚û§ Only supports constant discrete values
  ‚û§ Works with all data types               ‚û§ Only works with int, char, enum
  ‚û§ Better for complex conditions           ‚û§ Cleaner for menu/options/single variable match

------------------------------------------------------------
üìå SYNTAX of switch-case (line by line)

    switch (expression) {
        case constant1:
            // code block 1
            break;

        case constant2:
            // code block 2
            break;

        ...

        default:
            // default code block (optional)
    }

------------------------------------------------------------
üìå KEYWORDS EXPLAINED:

üîπ switch(expression):
    ‚û§ Evaluates `expression` (must be int, char, enum).

üîπ case constant:
    ‚û§ Compares the value of expression to this constant.
    ‚û§ Executes if match found.

üîπ break;
    ‚û§ Exits the switch block.
    ‚û§ If omitted, control falls through to the next case (‚ö†Ô∏è fall-through behavior).

üîπ default:
    ‚û§ Executes if no case matches.
    ‚û§ Optional but useful as a fallback.

------------------------------------------------------------
üìå FLOWCHART:

                 +--------------------+
                 | switch(expression) |
                 +--------------------+
                   |         |       |
              +----+         |       +----+
              |              |            |
       +------------+  +-------------+ ... +--------------+
       |  case 1 ?  |  |  case 2 ?   |     |  case n ?    |
       +------------+  +-------------+     +--------------+
            |                 |                   |
          True              True                True
            |                 |                   |
      +-----------+    +-----------+       +-------------+
      | Block 1   |    | Block 2   |  ...  | Block n     |
      +-----------+    +-----------+       +-------------+
            |                 |                   |
          break             break               break
            ‚Üì                 ‚Üì                   ‚Üì
       +---------------------------------------------+
       |              Exit from switch               |
       +---------------------------------------------+
            ‚Üë
            |
         False (from all)
            |
        +----------+
        | default  |
        |  block   |
        +----------+
             |
           break
             ‚Üì
      +-------------------+
      | Exit from switch  |
      +-------------------+

------------------------------------------------------------
üíª EXAMPLE 1: switch WITH break (recommended)

#include <stdio.h>
int main() {
    int choice;
    printf("Enter a number (1-3): ");
    scanf("%d", &choice);

    switch (choice) {
        case 1:
            printf("You chose One\n");
            break;
        case 2:
            printf("You chose Two\n");
            break;
        case 3:
            printf("You chose Three\n");
            break;
        default:
            printf("Invalid choice\n");
    }

    return 0;
}

------------------------------------------------------------
üíª EXAMPLE 2: switch WITHOUT break (fall-through behavior)

#include <stdio.h>
int main() {
    int choice = 2;

    switch (choice) {
        case 1:
            printf("One\n");
        case 2:
            printf("Two\n");
        case 3:
            printf("Three\n");
        default:
            printf("Default Case\n");
    }

    return 0;
}

üîé OUTPUT:
    Two
    Three
    Default Case

üéØ Explanation:
    ‚û§ Since there is no `break`, all cases from the match point (case 2) are executed.

------------------------------------------------------------*/
````

````python
/*------------------------------------------------------------
üìò WHILE LOOP IN C ‚Äî Complete Theory and Syntax
-------------------------------------------------------------

üìå WHAT IS A WHILE LOOP?
‚û§ A `while` loop is an **entry-controlled loop** in C programming.
‚û§ It **executes a block of code repeatedly** as long as the given condition is true.
‚û§ It's called a **top-checking loop** because the condition is checked *before* each iteration.

------------------------------------------------------------
üìå GENERAL SYNTAX:

    while (condition) {
        // code block to be repeated
    }

üîπ condition:
    ‚û§ Must return TRUE (non-zero) or FALSE (zero).
    ‚û§ The loop continues as long as condition is TRUE.
    ‚û§ The condition is checked before each iteration.

------------------------------------------------------------
üìå FLOW OF CONTROL in WHILE LOOP

               +------------------+
               |  condition true? |
               +--------+---------+
                        |
                   Yes  ‚Üì
              +------------------+
              |  Execute block   |
              +------------------+
                        |
                   Go back ‚Üë
                        |
               +------------------+
               |  condition false?|
               +--------+---------+
                        ‚Üì
                  Exit the loop

------------------------------------------------------------
üìå FLOWCHART (Copy-paste ready for comments)

              +----------------------+
              |  while (condition)   |
              +----------+-----------+
                         |
                   True  ‚Üì
              +------------------+
              | Execute block    |
              +------------------+
                         |
                         ‚îî‚îÄ‚îÄ‚îÄ‚Üí (go back to condition)
                         |
                      False ‚Üì
                 +----------------+
                 | Exit the loop  |
                 +----------------+

------------------------------------------------------------
üìå IMPORTANT: INFINITE LOOP

üî¥ If the condition never becomes false (like when counter is never updated), the loop becomes an infinite loop.

‚ö†Ô∏è Example:

    int i = 1;
    while (i <= 5) {
        printf("%d\n", i);
        // forgot to update i ‚û§ causes infinite loop
    }

‚úÖ Fix:

    int i = 1;
    while (i <= 5) {
        printf("%d\n", i);
        i++;  // condition gets updated ‚û§ finite loop
    }

------------------------------------------------------------
üìå WHEN TO USE WHILE LOOP?

‚úÖ Use `while` loop when:
  ‚û§ The number of iterations is **unknown** in advance.
  ‚û§ You want to **run the loop as long as a condition is true**.
  ‚û§ Perfect for user input until exit, reading files, sensors, etc.

------------------------------------------------------------
üíª PROGRAM: Print numbers from 1 to 5 using while loop

#include <stdio.h>
int main() {
    int i = 1;  // Initialization

    while (i <= 5) {  // Condition check before each loop
        printf("%d\n", i);  // Print current value
        i++;  // Update step ‚û§ move to next number
    }

    return 0;
}

üîç OUTPUT:
    1
    2
    3
    4
    5

üìå LOOP WORKING STEP BY STEP:
    i = 1 ‚û§ condition true ‚û§ print 1 ‚û§ i becomes 2  
    i = 2 ‚û§ condition true ‚û§ print 2 ‚û§ i becomes 3  
    i = 3 ‚û§ condition true ‚û§ print 3 ‚û§ i becomes 4  
    i = 4 ‚û§ condition true ‚û§ print 4 ‚û§ i becomes 5  
    i = 5 ‚û§ condition true ‚û§ print 5 ‚û§ i becomes 6  
    i = 6 ‚û§ condition false ‚û§ loop stops

------------------------------------------------------------*/
````

````python
/*------------------------------------------------------------
üìò DO...WHILE LOOP IN C ‚Äî Full Theory and Explanation
-------------------------------------------------------------

üìå WHAT IS A DO...WHILE LOOP?

‚û§ The `do...while` loop is a **bottom-checking loop** or **exit-controlled loop** in C.
‚û§ The loop body is executed **at least once**, even if the condition is false in the beginning.
‚û§ The condition is tested **after** the execution of the loop body.

------------------------------------------------------------
üìå SYNTAX:

    do {
        // loop body: statement(s)
    } while (condition);

------------------------------------------------------------
üìå FLOW OF CONTROL:

    1. Loop body is executed first.
    2. Then the condition is tested.
    3. If condition is TRUE, loop repeats.
    4. If FALSE, loop exits.

------------------------------------------------------------
üìå FLOWCHART (Text style for copy-paste in code)

              +------------------+
              |  Execute block   |
              +--------+---------+
                       |
               +-------v--------+
               |  condition?    |
               +---+--------+---+
                   |        |
                 True      False
                   |         ‚Üì
                   +---------+
                   |
           (Go back to block)

------------------------------------------------------------
üìå COMPARISON: while vs do...while

  Feature              | while                     | do...while
  ---------------------|----------------------------|----------------------------
  Condition checked    | Before loop starts         | After first iteration
  Loop type            | Entry-controlled loop      | Exit-controlled loop
  Loop body executes   | 0 or more times            | At least once
  Best used when       | Condition may not be true  | You want block to run once minimum

------------------------------------------------------------
üìå WHEN TO USE `do...while`?

‚úÖ Use `do...while` when:
  ‚û§ You want the code to **execute at least once**.
  ‚û§ Useful in **menu-driven programs**, **user input validation**, **games**, etc.

------------------------------------------------------------
üíª PROGRAM: Simple menu-driven calculator using do...while

#include <stdio.h>
int main() {
    int choice;
    int a = 10, b = 5;

    do {
        printf("\n--- MENU ---\n");
        printf("1. Add\n");
        printf("2. Subtract\n");
        printf("3. Multiply\n");
        printf("4. Divide\n");
        printf("5. Exit\n");
        printf("Enter your choice: ");
        scanf("%d", &choice);

        switch (choice) {
            case 1:
                printf("Result: %d\n", a + b);
                break;
            case 2:
                printf("Result: %d\n", a - b);
                break;
            case 3:
                printf("Result: %d\n", a * b);
                break;
            case 4:
                if (b != 0)
                    printf("Result: %.2f\n", (float)a / b);
                else
                    printf("Division by zero error!\n");
                break;
            case 5:
                printf("Exiting...\n");
                break;
            default:
                printf("Invalid choice! Try again.\n");
        }

    } while (choice != 5);  // loop runs until user enters 5

    return 0;
}

------------------------------------------------------------
üìå HOW THIS LOOP WORKS:

    üîπ Loop runs and shows menu (even if choice is 5 initially).
    üîπ User inputs a choice.
    üîπ If choice != 5, loop repeats.
    üîπ Ensures menu is shown at least once (even if invalid input or exit command).

------------------------------------------------------------
üìå TAKEAWAY:

‚úÖ Use `while` when you *may not* need the loop to run at all.  
‚úÖ Use `do...while` when you *always need one execution* before checking the condition.

------------------------------------------------------------*/
````
