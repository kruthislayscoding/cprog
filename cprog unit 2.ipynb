````python
//DANGLING ELSE AND COMPARING FLOAT NUMBERS USING EPSILON AND FABS()
/*
😵 Dangling else:
   - Happens when you write nested `if` without proper braces `{}`.
   - The compiler always associates the `else` with the nearest unmatched `if`.

🛑 BAD (Ambiguous):
    if (x > 0)
        if (x < 10)
            printf("Between 1 and 9");
        else
            printf("Not greater than 0");

✅ GOOD (Clear and intended):
    if (x > 0) {
        if (x < 10)
            printf("Between 1 and 9");
    } else {
        printf("Not greater than 0");
    }
*/
/*
🎯 Problem: Floating-point values can't be compared using `==` due to precision errors.
📌 Solution: Use an acceptable difference range (called EPSILON) and `fabs()` from <math.h>

    Example: To check if a and b are "almost equal":
        if (fabs(a - b) < EPSILON)

    - `fabs()` gives absolute value of float/double.
    - `EPSILON` is a small value like 0.000001
*/

#include <stdio.h>
#include <math.h>

#define EPSILON 0.000001

int main() {
    float a = 0.1f * 3;
    float b = 0.3f;

    printf("a = %.20f\n", a);  // just to show difference
    printf("b = %.20f\n", b);

    if (fabs(a - b) < EPSILON) {
        printf("✅ a and b are approximately equal.\n");
    } else {
        printf("❌ a and b are NOT equal.\n");
    }

    return 0;
}

````

````python
/*------------------------------------------------------------
  📘 SWITCH-CASE STATEMENT IN C — Full Explanation
-------------------------------------------------------------

📌 Purpose:
   ➤ Used for multi-way branching when comparing a variable to many constant values.
   ➤ Works **only with integral or enumerated types** (int, char, enum — not float/double).

------------------------------------------------------------
📚 COMPARISON: switch-case vs if-else ladder

  if-else ladder:                            switch-case:
  ----------------                          ------------------
  ➤ Supports ranges, logic ops              ➤ Only supports constant discrete values
  ➤ Works with all data types               ➤ Only works with int, char, enum
  ➤ Better for complex conditions           ➤ Cleaner for menu/options/single variable match

------------------------------------------------------------
📌 SYNTAX of switch-case (line by line)

    switch (expression) {
        case constant1:
            // code block 1
            break;

        case constant2:
            // code block 2
            break;

        ...

        default:
            // default code block (optional)
    }

------------------------------------------------------------
📌 KEYWORDS EXPLAINED:

🔹 switch(expression):
    ➤ Evaluates `expression` (must be int, char, enum).

🔹 case constant:
    ➤ Compares the value of expression to this constant.
    ➤ Executes if match found.

🔹 break;
    ➤ Exits the switch block.
    ➤ If omitted, control falls through to the next case (⚠️ fall-through behavior).

🔹 default:
    ➤ Executes if no case matches.
    ➤ Optional but useful as a fallback.

------------------------------------------------------------
📌 FLOWCHART:

                 +--------------------+
                 | switch(expression) |
                 +--------------------+
                   |         |       |
              +----+         |       +----+
              |              |            |
       +------------+  +-------------+ ... +--------------+
       |  case 1 ?  |  |  case 2 ?   |     |  case n ?    |
       +------------+  +-------------+     +--------------+
            |                 |                   |
          True              True                True
            |                 |                   |
      +-----------+    +-----------+       +-------------+
      | Block 1   |    | Block 2   |  ...  | Block n     |
      +-----------+    +-----------+       +-------------+
            |                 |                   |
          break             break               break
            ↓                 ↓                   ↓
       +---------------------------------------------+
       |              Exit from switch               |
       +---------------------------------------------+
            ↑
            |
         False (from all)
            |
        +----------+
        | default  |
        |  block   |
        +----------+
             |
           break
             ↓
      +-------------------+
      | Exit from switch  |
      +-------------------+

------------------------------------------------------------
💻 EXAMPLE 1: switch WITH break (recommended)

#include <stdio.h>
int main() {
    int choice;
    printf("Enter a number (1-3): ");
    scanf("%d", &choice);

    switch (choice) {
        case 1:
            printf("You chose One\n");
            break;
        case 2:
            printf("You chose Two\n");
            break;
        case 3:
            printf("You chose Three\n");
            break;
        default:
            printf("Invalid choice\n");
    }

    return 0;
}

------------------------------------------------------------
💻 EXAMPLE 2: switch WITHOUT break (fall-through behavior)

#include <stdio.h>
int main() {
    int choice = 2;

    switch (choice) {
        case 1:
            printf("One\n");
        case 2:
            printf("Two\n");
        case 3:
            printf("Three\n");
        default:
            printf("Default Case\n");
    }

    return 0;
}

🔎 OUTPUT:
    Two
    Three
    Default Case

🎯 Explanation:
    ➤ Since there is no `break`, all cases from the match point (case 2) are executed.

------------------------------------------------------------*/
````

````python
/*------------------------------------------------------------
📘 WHILE LOOP IN C — Complete Theory and Syntax
-------------------------------------------------------------

📌 WHAT IS A WHILE LOOP?
➤ A `while` loop is an **entry-controlled loop** in C programming.
➤ It **executes a block of code repeatedly** as long as the given condition is true.
➤ It's called a **top-checking loop** because the condition is checked *before* each iteration.

------------------------------------------------------------
📌 GENERAL SYNTAX:

    while (condition) {
        // code block to be repeated
    }

🔹 condition:
    ➤ Must return TRUE (non-zero) or FALSE (zero).
    ➤ The loop continues as long as condition is TRUE.
    ➤ The condition is checked before each iteration.

------------------------------------------------------------
📌 FLOW OF CONTROL in WHILE LOOP

               +------------------+
               |  condition true? |
               +--------+---------+
                        |
                   Yes  ↓
              +------------------+
              |  Execute block   |
              +------------------+
                        |
                   Go back ↑
                        |
               +------------------+
               |  condition false?|
               +--------+---------+
                        ↓
                  Exit the loop

------------------------------------------------------------
📌 FLOWCHART (Copy-paste ready for comments)

              +----------------------+
              |  while (condition)   |
              +----------+-----------+
                         |
                   True  ↓
              +------------------+
              | Execute block    |
              +------------------+
                         |
                         └───→ (go back to condition)
                         |
                      False ↓
                 +----------------+
                 | Exit the loop  |
                 +----------------+

------------------------------------------------------------
📌 IMPORTANT: INFINITE LOOP

🔴 If the condition never becomes false (like when counter is never updated), the loop becomes an infinite loop.

⚠️ Example:

    int i = 1;
    while (i <= 5) {
        printf("%d\n", i);
        // forgot to update i ➤ causes infinite loop
    }

✅ Fix:

    int i = 1;
    while (i <= 5) {
        printf("%d\n", i);
        i++;  // condition gets updated ➤ finite loop
    }

------------------------------------------------------------
📌 WHEN TO USE WHILE LOOP?

✅ Use `while` loop when:
  ➤ The number of iterations is **unknown** in advance.
  ➤ You want to **run the loop as long as a condition is true**.
  ➤ Perfect for user input until exit, reading files, sensors, etc.

------------------------------------------------------------
💻 PROGRAM: Print numbers from 1 to 5 using while loop

#include <stdio.h>
int main() {
    int i = 1;  // Initialization

    while (i <= 5) {  // Condition check before each loop
        printf("%d\n", i);  // Print current value
        i++;  // Update step ➤ move to next number
    }

    return 0;
}

🔍 OUTPUT:
    1
    2
    3
    4
    5

📌 LOOP WORKING STEP BY STEP:
    i = 1 ➤ condition true ➤ print 1 ➤ i becomes 2  
    i = 2 ➤ condition true ➤ print 2 ➤ i becomes 3  
    i = 3 ➤ condition true ➤ print 3 ➤ i becomes 4  
    i = 4 ➤ condition true ➤ print 4 ➤ i becomes 5  
    i = 5 ➤ condition true ➤ print 5 ➤ i becomes 6  
    i = 6 ➤ condition false ➤ loop stops

------------------------------------------------------------*/
````

````python
/*------------------------------------------------------------
📘 DO...WHILE LOOP IN C — Full Theory and Explanation
-------------------------------------------------------------

📌 WHAT IS A DO...WHILE LOOP?

➤ The `do...while` loop is a **bottom-checking loop** or **exit-controlled loop** in C.
➤ The loop body is executed **at least once**, even if the condition is false in the beginning.
➤ The condition is tested **after** the execution of the loop body.

------------------------------------------------------------
📌 SYNTAX:

    do {
        // loop body: statement(s)
    } while (condition);

------------------------------------------------------------
📌 FLOW OF CONTROL:

    1. Loop body is executed first.
    2. Then the condition is tested.
    3. If condition is TRUE, loop repeats.
    4. If FALSE, loop exits.

------------------------------------------------------------
📌 FLOWCHART (Text style for copy-paste in code)

              +------------------+
              |  Execute block   |
              +--------+---------+
                       |
               +-------v--------+
               |  condition?    |
               +---+--------+---+
                   |        |
                 True      False
                   |         ↓
                   +---------+
                   |
           (Go back to block)

------------------------------------------------------------
📌 COMPARISON: while vs do...while

  Feature              | while                     | do...while
  ---------------------|----------------------------|----------------------------
  Condition checked    | Before loop starts         | After first iteration
  Loop type            | Entry-controlled loop      | Exit-controlled loop
  Loop body executes   | 0 or more times            | At least once
  Best used when       | Condition may not be true  | You want block to run once minimum

------------------------------------------------------------
📌 WHEN TO USE `do...while`?

✅ Use `do...while` when:
  ➤ You want the code to **execute at least once**.
  ➤ Useful in **menu-driven programs**, **user input validation**, **games**, etc.

------------------------------------------------------------
💻 PROGRAM: Simple menu-driven calculator using do...while

#include <stdio.h>
int main() {
    int choice;
    int a = 10, b = 5;

    do {
        printf("\n--- MENU ---\n");
        printf("1. Add\n");
        printf("2. Subtract\n");
        printf("3. Multiply\n");
        printf("4. Divide\n");
        printf("5. Exit\n");
        printf("Enter your choice: ");
        scanf("%d", &choice);

        switch (choice) {
            case 1:
                printf("Result: %d\n", a + b);
                break;
            case 2:
                printf("Result: %d\n", a - b);
                break;
            case 3:
                printf("Result: %d\n", a * b);
                break;
            case 4:
                if (b != 0)
                    printf("Result: %.2f\n", (float)a / b);
                else
                    printf("Division by zero error!\n");
                break;
            case 5:
                printf("Exiting...\n");
                break;
            default:
                printf("Invalid choice! Try again.\n");
        }

    } while (choice != 5);  // loop runs until user enters 5

    return 0;
}

------------------------------------------------------------
📌 HOW THIS LOOP WORKS:

    🔹 Loop runs and shows menu (even if choice is 5 initially).
    🔹 User inputs a choice.
    🔹 If choice != 5, loop repeats.
    🔹 Ensures menu is shown at least once (even if invalid input or exit command).

------------------------------------------------------------
📌 TAKEAWAY:

✅ Use `while` when you *may not* need the loop to run at all.  
✅ Use `do...while` when you *always need one execution* before checking the condition.

------------------------------------------------------------*/
````

````python

/*------------------------------------------------------------
📘 FOR LOOP IN C — Full Theory, Syntax, Flowchart & Program
-------------------------------------------------------------

📌 WHAT IS A FOR LOOP?

➤ A `for` loop is a **definite loop** in C.
➤ It is used when the **number of iterations is known** in advance.
➤ It is an **entry-controlled loop**, meaning the condition is checked before entering the loop.
➤ Syntax is compact — all loop control steps (init, condition, increment) are in one line.

------------------------------------------------------------
📌 SYNTAX OF FOR LOOP:

    for (initialization; condition; increment/decrement) {
        // loop body: statements to execute repeatedly
    }

🔹 initialization     ➤ executed once at the beginning
🔹 condition          ➤ evaluated before every loop iteration
🔹 increment/decrement ➤ executed at the end of every iteration

------------------------------------------------------------
📌 FLOW OF CONTROL:

1. Initialization → Run once
2. Check condition:
    🔹 If TRUE → execute loop body
    🔹 If FALSE → exit loop
3. Execute loop body
4. Increment/decrement
5. Go back to step 2

------------------------------------------------------------
📌 FLOWCHART (Text-based and copy-paste friendly)

               +--------------------+
               | Initialization     |
               +---------+----------+
                         ↓
               +--------------------+
               | Condition true?    |
               +----+-----------+---+
                    |           |
                   Yes         No
                    ↓           ↓
              +-----------+    +------------------+
              | Loop Body |    | Exit the Loop    |
              +-----+-----+    +------------------+
                    |
                    ↓
              +-----------+
              | Increment |
              +-----------+
                    |
              (go back to condition)

------------------------------------------------------------
📌 EXAMPLE PROGRAM — Print numbers from 1 to 5

#include <stdio.h>
int main() {
    for (int i = 1; i <= 5; i++) {
        printf("%d\n", i);
    }
    return 0;
}

📌 STEP-BY-STEP EXPLANATION:
1. int i = 1 → initialized once
2. i <= 5? → TRUE → enter loop
3. print i → prints 1
4. i++ → i becomes 2 → go to condition
5. Repeat steps 2–4 until i = 6 → FALSE → exit loop

------------------------------------------------------------
📌 EMPTY CONDITIONS EXAMPLE:

for(;;) {
    // infinite loop (no condition = always true)
    printf("Hello\n");
}

➤ All parts of for loop are optional.
➤ `for(;;)` is equivalent to an infinite loop like `while(1)`.

------------------------------------------------------------
📌 WHEN TO USE FOR LOOP?

✅ Use `for` when:
- Number of iterations is **known**.
- You want compact loop control in one line.
- Useful in counting, printing patterns, arrays, etc.

------------------------------------------------------------
📌 COMPARISON: for vs while vs do...while

  Feature           | for                      | while                     | do...while
  ------------------|---------------------------|----------------------------|----------------------------
  Control type      | Entry-controlled          | Entry-controlled          | Exit-controlled
  Use case          | Known iteration count     | Unknown iterations        | At least one execution needed
  Syntax            | Compact, one-line control | More spread out           | Loop body then condition
  Executes body     | 0 or more times           | 0 or more times           | 1 or more times

------------------------------------------------------------*/
//KEY POINTS TO REMEMBER ABOUT FOR LOOP:
1: IN A FOR LOOP,ANY OR ALL EXPRESSIONS CAN BE OMITTED IN CASE ALL OMITTED THEN THERE MUST BE TWO SEMICOLONS IN THE FOR STATEMENT
2:THERE MUST BE NO SEMICOLON AFTER A FOR STATEMENT, IF YOU DO THAT, SURE TO GET UNEXPECTED RESULT:
CONSIDER EX: INT MAIN(){ int i; for(i=0;i<10;i++); printf("%d",i); return 0;}
//in this code the loop initialises i to 0 and increments its value, since a ; is placed after the loop it means that loop does not contain any 
//statement so even if the condition is true, no statement is executed, THE LOOP CONTINUES TILL i becomes 10 and the moment i=10, the statement
//following the for loop is executed and the value of i(10) is printed on screen.
3:WHEN WE PLACE SEMICOLON AFTER FOR LOOP THEN COMPUTER WONT GENERATE ANY ERROR MSG RATHER IT WILL TREAT THE STATEMENT AS A NULL STATEMENT
USUALLY SUCH TYPE OF NULL STATEMENTS IS USED TO GENERATE TIME DELAYS 
4:MULTIPLE INITIALISATIONS MUST BE SEPERATED WITH A COMMA OPERATOR:
#include<stdio.h>
int main()
{
    int i,num;
    for(i=0;sum=0;i<10;i++)
           sum+=i;
    printf("%d",sum);
    return 0;
}//as shown here is perfect
5:if there is no initialisation to be done, then the initialisation statement can be skipped by giving only a semicolon
#include<stdio.h>
int main(){
    int i=0;
    for(;i<10;i++)
       pf("%d",i);
    return 0;
}
6:NEVER USE A FLOAT VAR AS LOOP CONTROL VAR AS IT LEADS TO APPROXIMATIONS,MAY LEAD TO IMPRECISE VALUES, AND INFINITE LOOP
````
