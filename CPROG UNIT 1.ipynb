```python
//AREA OF TRIANGLE USING HERON'S FORMULA
#include<stdio.h>
#include<math.h>
#include<conio.h>
int main()
{
    float a,b,c,area,S;
    printf('\n enter the length of three sides of triangle:');
    scanf('%f %f %f',&a,&b,&c);
    S=(a+b+c)/2;
//sqrt is a mathematical fn in math.h
    area=sqrt(S*(S-a)*(S-b)*(S-c));
    printf('\n area=%f',area);
    return 0;
}
//OUTPUT
enter the length of three sides of triangle: 12 16 20
area=96
```

```python
//VVIMP PGM TO SHOW FORMATTING SPECIFIERS
#include <stdio.h>

int main(void) 
{ 
    int a = 234, b = -234, c = 54321;

    // Prints c with a minimum width of 2 (not applicable as 54321 already has more digits)
    printf("%2d\n", c); 

    // Prints c with a minimum width of 10 but .2d is not a valid precision specifier for integers
    // The effect would be similar to just "%10d" (right-aligned)
    printf("%10.2d\n", c); 

    // Left-aligns c within a width of 10 spaces
    printf("%-10.2d\n", c); 

    // Left-aligns a within a width of 7
    printf("%-7d\n", a); 

    // Right-aligns a within a width of 7, fills empty spaces with '0'
    printf("%07.2d\n", a); 

    // Right-aligns a within a width of 7, fills empty spaces with '0'
    printf("%07d\n", a); 

    // '+': Forces sign display, '-': Left-align, '0': Fill empty spaces with 0, width=9, precision=4
    // However, '-0' is invalid together, so left-align (-) dominates.
    printf("%+0-9.4d\n", a); 

    // '+': Forces sign display, '0': Fill empty spaces with 0, width=9, precision=4
    printf("%+09.4d\n", a); 

    // '+': Forces sign display, '0': Fill empty spaces with 0, width=7
    printf("%+07d\n", a); 

    // '+': Forces sign display, '0': Fill empty spaces with 0, width=7, precision=4
    printf("%+07.4d\n", a); 

    // '+': Forces sign display, '-': Left-align, width=7, precision=4
    printf("%+-07.4d\n", a); 

    // Left-aligns b within a width of 8
    printf("%-08d\n", b); 

    // Left-aligns b within a width of 8 with .2d (not affecting display)
    printf("%-08.2d\n", b); 

    // Left-aligns b within a width of 8, ensures at least 4 digits with leading zeros if needed
    printf("%-8.4d\n", b); 

    return 0; 
}  

//OUTPUT
54321
      54321
54321     
234    
0000234
0000234
+0234    
+00234
+0000234
+00234
+00234  
-234    
-234    
-0234   

```
